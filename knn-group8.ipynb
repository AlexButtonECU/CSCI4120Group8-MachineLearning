```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
from sklearn import *

url = 'https://raw.githubusercontent.com/ruiwu1990/CSCI_4120/master/KNN/iris.data'

def main():    
	yAxis = []
	for m in range(20):
		accuracy = []      
		k = m + 1
		for i in range(5):
			
			iris = load_iris()
			x = iris.data
			y =iris.target
			x1, x2, y1, y2 = train_test_split(x, y, test_size=0.33, )
			model = KNeighborsClassifier(n_neighbors=k)
			model.fit(x1, y1)
			predicted = model.predict(x2)
			accuracy.append(accuracy_score(y2, predicted))
		yAxis.append(sum(accuracy)/5)
	plt.plot(range(k), yAxis)
	plt.xlim(0,20)
	plt.ylim(0.5,1.0)
	plt.xlabel("K")
	plt.ylabel("Accuracy")
	plt.title("Accuracy of KNN model over increasing K")
	
main()
```
